```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
it = pd.read_csv('/content/IT Salary Survey EU 2019.csv', sep =',')


it['Age'].hist(bins=13)

plt.show()
```

```python
it['Company business sector'].value_counts()[:10].plot.barh(color='purple')
```

```python
it['Your main technology / programming language'].value_counts().plot.bar(color='red', rot=60)
```

```python
fig = plt.figure(figsize=(15,5))

ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)

# Add a plot to each of the axes created above.
# Plot the histogram of the variable 'Age' with 13 bins.

ax1.hist(it['Age'], bins = 13)

# Plot the horizontal bars of the 10 most common company business sectors.
# The variable is 'Company business sector'.
# The color of the bars should purple.

ax2.barh(it['Company business sector'].value_counts()[:10].index, it['Company business sector'].value_counts()[:10], color = 'purple')

# Plot the main programming language that the data scientists use at work.
# The variable is 'Your main technology / programming language'.
# The color of the bars should red.

ax3.bar(it['Your main technology / programming language'].value_counts().index, it['Your main technology / programming language'].value_counts(), color ='red')

ax3.tick_params(axis='x', labelrotation = 60)

fig.tight_layout()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.pie(x=list(it['Gender'].value_counts()),
        labels = ['Male', 'Female'],
        colors = ['black','purple'],
        explode = [0,0.1])
fig.show()
```

```python
ax1 = fig.add_subplot(1,1,1)
ax1.pie(x=list(it['Gender'].value_counts()), labels = ['Male','Female'], colors = ['black','purple'], explode = [0, 0.1])
fig.show()
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(1,1,1)

colors = {'Male': 'black', 'Female': 'purple'}

ax1.scatter(x=it['Yearly brutto salary (without bonus and stocks)'], y=it['Yearly bonus'], marker= '+', c=it['Gender'].map(colors))

ax1.set_title('Yearly brutto salary vs Yearly bonus')
ax1.set_xlabel('Yearly brutto salary')
ax1.set_ylabel('Yearly bonus')

plt.grid()
fig.show()
```

```python
import seaborn as sns
```

```python
sns.displot(data=it, x='Yearly brutto salary (without bonus and stocks)', kde= True, col= 'Gender')
```

```python
sns.displot(data=it, x='Yearly brutto salary (without bonus and stocks)', kind='ecdf', hue = 'Gender')
```

```python
it2 = it[it['Seniority level']== 'Middle']
sns.displot(data=it2, x='Yearly brutto salary (without bonus and stocks)', kind='ecdf', hue = 'Gender')
```

```python
sns.displot(data=it2, x='Yearly bonus', kind='ecdf', hue = 'Gender')
```

```python
sns.catplot(data=it, kind='violin', x='Seniority level', y='Yearly brutto salary (without bonus and stocks)',
            hue='Gender')
```

```python
it['Seniority level'].value_counts()
```

```python
sns.lmplot(x='Yearly brutto salary (without bonus and stocks)',
           y='Yearly bonus',
           data=it,
           hue = 'Gender',
           fit_reg= False,
           col= 'Seniority level')
```

```python
# We are now going to use JoinGrid from Seaborn to create a scatter plot of
# variables 'Yearly brutto salary (without bonus and stocks)' versus 'Yearly bonus',
# together with the associated Kernel density plots in the outer areas.
# We also want different colours for males and females.

# We want to focus our attention on the seniority levels 'Middle' and 'Senior'.
# Create 'it3', which is a sub sample of 'it' that includes all the data
# scientists with a seniority level (variable 'Seniority level') mentioned above
# using the .isin() method.


it3 = it[it['Seniority level'].isin(['Middle', 'Senior'])]

g= sns.JointGrid(data=it3, x='Yearly brutto salary (without bonus and stocks)',
                 y= 'Yearly bonus', hue ='Gender')
g= g.plot_joint(sns.scatterplot)
g=g.plot_marginals(sns.kdeplot, shade= True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
google = pd.read_csv('/content/Trudeau_Singh.csv', sep = ',')
```

```python
google['Date'] = pd.to_datetime(google['Week'])
google
```

```python
# Use matplotlib to plot the time series of google searches associated
# with Jagmeet Singh (variable 'Jagmeet Singh').
# Use the color 'mediumaquamarine'.
plt.plot(google['Date'],google['Jagmeet Singh'],color = 'mediumaquamarine')
plt.show()
```

```python

# We need the 'datetime' library to plot a vertical bar at a specific date.

import datetime as dt
fig, ax = plt.subplots()

ax.plot(google['Date'], google['justin trudeau'],':', color = 'darkred')
ax.plot(google['Date'], google['Jagmeet Singh'], '--', color = 'teal')
ax.axvline(dt.datetime(2021,9,19), linewidth= 4, color= 'lightgray', alpha= 0.7)

ax.text(dt.datetime(2021, 7, 25), 85, 'Election day', fontsize =12)

ax.legend(['Justin Trudeau', ' Jagmeet Singh'])

ax.tick_params(axis='x', labelrotation=40)

for spine in ['top', 'right']:
          ax.spines[spine].set_visible(False)
plt.show()
```
